```python
import numpy as np
import math
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
import numpy as np
X_iono = np.genfromtxt("ionosphere.txt", delimiter=",", usecols=np.arange(34))
y_iono = np.genfromtxt("ionosphere.txt", delimiter=",", usecols=34, dtype='int')

```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train_iris, X_test_iris, y_train_iris, y_test_iris = train_test_split(iris['data'], 
                                                                        iris['target'], random_state=1103)
```

```python
X_train_iono, X_test_iono, y_train_iono, y_test_iono = train_test_split(X_iono, y_iono, random_state=1103)
```

```python

# To calculate the Euclidean dist. of two rows/vectors of data.
def euclidean_distance(row1, row2):
    dist = 0.0
    length = len(row1)
    for i in range(length):
        dist += (row1[i] - row2[i])**2 
    return math.sqrt(dist)
```

```python
# Testing Euclidean dist calculation
row1 = [1, 2]
row2 = [3, 4]

print(euclidean_distance(row1, row2))
```

```python
import math

def nearest_neighbor(array, test_data, train_label):
    test_data_shape = test_data.shape[0]
    predictions = np.zeros(test_data_shape, dtype=int)
    for i in range(test_data_shape):
        best = 9999999
        input = test_data[i]
        for j in range(array.shape[0]):
            current = euclidean_distance(input, array[j])
            if current < best:
                best = current
                best_place = j
        predictions[i] = train_label[best_place]
    return predictions
    
```

```python
# Running nearest_neighbor on iris dataset, this gives us an array with
# our predictions.
prediction = nearest_neighbor(X_train_iris, X_test_iris, y_train_iris)
```

```python
# Now lets loop through our predictions and check them against labels.

errors = 0
iris_test_shape = X_test_iris.shape[0]
# Loop through the size of dataset and check whether prediction is same
# as 
for i in range(iris_test_shape):
    if prediction[i] != y_test_iris[i]:
        errors = errors + 1
error_rate = errors/iris_test_shape

print("The number of errors is ", errors)
print("the test error rate for Nearest Neighbour applied to the iris dataset; ", error_rate)
        
```

```python
# Now for ionosphere database we get our predictions in the same way.
prediction = nearest_neighbor(X_train_iono, X_test_iono, y_train_iono)


# Just checking, by eye, the prediction length is the same as test set label length
print("Preditction length is : ", len(prediction))
print("labels of test length is : ", len(y_test_iono))
```

```python
# Now lets find our error rate by checking our prediction against labels

iono_test_shape = y_test_iono.shape[0]

errors = 0
for i in range(iono_test_shape):
    
    if prediction[i] != y_test_iono[i]:
        errors = errors + 1

error_rate2 = errors/iono_test_shape

print("The number of errors is ", errors)
print("the test error rate for Nearest Neighbour applied to ionosphere.txt; ", error_rate2)
```

```python
import math
def conf_pred_distance(train_data, train_label):
    train_data_shape = train_data.shape[0]
    tds_range = range(train_data_shape)
    best_same, best_other = [[math.inf for i in tds_range] for j in range(2)]
    
    for i in (range(train_data_shape)):
        
        for j in range(i + 1, train_data_shape):
            
            distance = euclidean_distance(train_data[j], train_data[i])
            
            if (distance < best_same[i]) and (train_label[j] == train_label[i]):
                
                best_same[i] = distance
            
            if (distance < best_same[j]) and (train_label[j] == train_label[i]):
                
                best_same[j] = distance
            
            if (train_label[j] != train_label[i]):
                
                if (distance < best_other[i]):
                    
                    best_other[i] = distance
                
                if (distance < best_other[j]):
                    
                    best_other[j] = distance
    
    return (best_same, best_other)

    
    
```
